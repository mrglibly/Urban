```python
import numpy as np
import time
```

```python
class User:
    
    def __init__(self, nickname: str, password: int, age: int):
        self.nickname = nickname
        self.password = password
        self.age = age
        
    def __str__(self):
        return (f'This is {self.nickname}, {self.age}, the User!')
    
    def __repr__(self):
        return (f'{self.nickname}')
    
    def __contains__(self, set):
        return self in set
            
class Video:
    
    def __init__(self, title, duration, time_now = 0, adult_mode = False):
        self.titel = title
        self.duration = duration
        self.time_now = time_now
        self.adult_mode = adult_mode

class UrTube:   
    # videos = []
    # users = []
    
    def __init__(self, users=[], videos=[], current_user: User = None):
        self.users = users
        self.videos = videos
        self.current_user = current_user
    
    def add(self, *args:Video):
        self.args = args
        
        for i in args:
            if len(self.videos) == 0 or i.titel not in np.array(self.videos)[0]:
                self.videos.append([i.titel,i.duration,i.time_now, i.adult_mode])
            else:
                continue
    
    def log_in(self, nickname, password):
        self.nickname = nickname
        self.password = password
        
        if self.nickname in np.array(self.users)[:,0] and str(hash(self.password)) in np.array(self.users)[:,1]:
            self.current_user = self.nickname
            print('Nick is logged_in!))...')
        else: 
            print('Nick not accepted!')
    
    def register(self, nick, passw, age):
        self.nick = nick
        self.passw = passw
        self.age = age
        
        if len(self.users) == 0 or self.nick not in np.array(self.users):
            self.users.append([self.nick, hash(self.passw), self.age])
            print(f'Пользователь {self.nick} зарегистрирован!')
            self.current_user = self.nick
        else:
            print(f'Пользователь {self.nick} уже существует!')
    
    def log_out(self):
        self.current_user = None
    
    def get_videos(self, word):
        filtered=[]
        for i in self.videos:
            if word.lower() in i[0].lower():
                filtered.append(i)
        return filtered
    
    def watch_video(self, title):
        self.title = title
        
        if self.current_user == None:
            print("Войдите в аккаунт, чтобы смотреть видео")
            return None
        
        for i in self.videos:
            if title == i[0]:
                d = i[1]
                adm = i[3]
                break
            else:
                continue
        
        for i in self.users:
            if self.current_user == i[0]:
                age_ = i[2]
                break
            else:
                continue
        
        if self.title in np.array(self.videos)[:,0]:
            if age_ < 18 and adm == True:
                print("Вам нет 18 лет, пожалуйста покиньте страницу!")
            else:
                print(f'Начинаем просмотр видео "{self.title}:"')
                for i in range(d, 0, -1):
                    print(i)
                    time.sleep(1)  # Ожидание 1 секунду перед следующим числом
                print("Конец видео!")    
```
